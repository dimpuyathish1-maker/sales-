```python
columns_needed = [ "ORDERNUMBER", "ORDERDATE", "STATUS", "QUANTITYORDERED", "PRICEEACH","SALES", "CONTACTFIRSTNAME", "CONTACTLASTNAME", "DEALSIZE","TERRITORY", "CITY", "STATE", "COUNTRY"]
df = df[columns_needed]
```

```python
duplicates_order = df[df.duplicated(subset="ORDERNUMBER", keep=False)]

print(duplicates_order)
```

```python
df = df.drop_duplicates(subset="ORDERNUMBER", keep="first")
#Keeps the first occurrence of the duplicate and removes the rest. This is the most common choice
```

```python
print(df["ORDERNUMBER"].duplicated().sum()) 
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], errors='coerce')
```

```python
text_columns = ['STATUS', 'CONTACTFIRSTNAME', 'CONTACTLASTNAME', 
             'DEALSIZE', 'TERRITORY', 'CITY', 'STATE', 'COUNTRY']

for columns in text_columns:
    df[columns] = df[columns].str.strip()
```

```python
df
```

```python
# Check for underscores,commas in all string columns
for columns in text_columns:
    underscores = df[columns].str.contains('_').sum()
    commas = df[columns].str.contains(',').sum()
    if underscores > 0 or commas > 0:
        print(f"{columns}: {underscores} rows with _, {commas} rows with ,")
```

```python
df = df.reset_index(drop=True)
```

```python
df['STATE'] = df['STATE'].fillna('NA')
df['TERRITORY'] = df['TERRITORY'].fillna('Unknown')
```

```python
# Saving the cleaned DataFrame
df.to_csv(r"C:\Users\DIMPU YATHISH\Downloads\cleaned_sales_data.csv", index=False)
```
